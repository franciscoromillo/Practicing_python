10. Crear una matriz de 10x10 con 1's en los bordes y 0 en el interior 

```python
#OPCIÓN 1
m = np.zeros((10,10))
m[0,:] = 1
m[-1,:] = 1
m[1:-1,0] = 1
m[1:-1,-1] = 1
print(m)
```

```python
#OPCIÓN 2
m2 = np.ones((10,10))
m2[1:-1, 1:-1]=0
print(m2)
```
