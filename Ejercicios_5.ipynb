3.5. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.

```python
import numpy as np
```

```python
df_pclass1 = df[['pclass', 'survived']].groupby('pclass').agg(np.sum)
df_pclass2 = df[['pclass', 'survived']].groupby('pclass').count()
df_pclass2.rename(columns={'survived':'total'}, inplace=True)
```

```python
df_pclass1
```

```python
df_pclass2
```

```python
df_pclass = pd.concat([df_pclass1, df_pclass2], axis=1)
df_pclass['pct'] = df_pclass['survived']/df_pclass['total']
df_pclass
```
